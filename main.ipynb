```python
from pathlib import Path

from constants import TARGET_TICKERS, INTERVAL, CALL_DELAY_SECONDS
from utils.get_max_available_share_price_data import get_max_available_share_price_data
from utils.print_header import print_header

print_header("RUN CONFIG")
print(f"Tickers: {', '.join(TARGET_TICKERS)}")
print(f"Trading interval: {INTERVAL}")
print(f"API call delay: {CALL_DELAY_SECONDS}(s)")

print_header("CALL LOG")


all_results: list[dict[str, object]] = []
for ticker in TARGET_TICKERS:
    csv_path, calls = get_max_available_share_price_data(
        target_ticker=ticker,
        interval=INTERVAL,
        delay_seconds=CALL_DELAY_SECONDS,
    )
    all_results.append(
        {
            "ticker": ticker,
            "csv_path": str(csv_path) if isinstance(csv_path, Path) else None,
            "calls": calls,
        }
    )
```
